This notebook demonstrates how business architecture and decision intelligence can be applied to complex operational environments using an airport operations scenario.

The focus is on identifying high-leverage decision areas rather than optimising individual processes in isolation.


## Business question

Which operational decisions within airport environments generate the highest commercial and customer impact, and where should data and decision intelligence capabilities be prioritised?


## Data

The dataset represents decision areas across multiple airports, including:
- Decision volume
- Operational delay impact
- Cost impact
- Customer impact
- Data quality and automation potential


```
import pandas as pd

df = pd.read_csv("../data/airport_decision_intelligence_dataset.csv")
df.head()
```

## Analytical approach

Decisions are assessed across financial impact, customer impact, and automation potential to identify where improved business architecture and analytics deliver the greatest leverage.


```
summary = df.groupby("decision_type").agg({
    "cost_impact_gbp":"mean",
    "customer_impact_score":"mean",
    "automation_potential":"mean",
    "daily_volume":"mean"
}).reset_index()

summary
```

```
summary["decision_leverage_score"] = (
    summary["cost_impact_gbp"].rank(pct=True) +
    summary["customer_impact_score"].rank(pct=True) +
    summary["automation_potential"].rank(pct=True)
)

summary.sort_values("decision_leverage_score", ascending=False)
```

## Key findings

- A small number of operational decisions drive disproportionate cost and customer impact.
- Decision leverage varies by decision type rather than airport.
- Business architecture clarity enables targeted data and analytics investment.
